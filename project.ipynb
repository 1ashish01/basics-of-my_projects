```python
import matplotlib.pyplot as plt
import numpy as np

# Define the productivity data
productivity_data = {
    "8 AM": 4.2,
    "12 PM": 3.5,
    "4 PM": 2.9,
    "8 PM": 3.1
}

# Create a bar chart
x = np.arange(len(productivity_data))
fig, ax = plt.subplots()
ax.bar(x, productivity_data.values(), align='center', alpha=0.5)
ax.set_xticks(x)
ax.set_xticklabels(productivity_data.keys())
ax.set_ylabel('Productivity Level')
ax.set_xlabel('Time of Day')
ax.set_title('Productivity Levels by Time of Day')

# Display the chart
plt.show()
```
