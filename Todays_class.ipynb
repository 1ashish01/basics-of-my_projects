```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=[1,2,3]
Y=[3,1,2]
plt.scatter(X,Y)
```

```python
slope_curr=0
intercept_curr=0
lr = 0.01
iterat = 1000
```

```python
for i in range(iterat):
    s1 = 0
    s2 = 0
    for j in range(len(X)):
        s1 += X[j] * (Y[j] - slope_curr*X[j] - intercept_curr)
        s2 += Y[j] - slope_curr*X[j] - intercept_curr
    slope_gradient = -(2/len(X)) * s1
    intercept_gradient = -(2/len(X)) * s2
    slope_curr -= LR * slope_gradient
    intercept_curr -= LR * intercept_gradient
    y_predicted = [slope_curr*x + intercept_curr for x in X]
    plt.plot(X, y_predicted)
plt.scatter(X, Y)
y_predicted = [slope_curr*x + intercept_curr for x in X]
plt.plot(X, y_predicted)
plt.show()
```
